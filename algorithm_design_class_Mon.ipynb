# # Brute-force method : 억지 기법


# 1. 0/1 배낭 채우기 문제(0/1 knapsack problem) :
  - 목적: 가능한 모든 경우의 조합을 시도하여 최대 가치를 찾음 (브루트포스 접근법).
  - 모든 가능한 부분 집합을 탐색하여 배낭 용량을 초과하지 않으면서 최대 가치를 계산
  - 시간복잡도: O(2^n)

```python
# subset은 n-비트의 이진수로 나타난 부분 집합을 의미. 출력된 각 리스트는 n개의 요소 중 어떤 것이 포함되었는지를 보여줌.
```

```python
def knapsack01_bruteforce(weights, values, W):
    # 전체 물건의 수
    n =len(weights)
    # 최대 가치
    best_value = 0
    # 부분 집한의 수는 2^n개 이므로, i에서 0부터  2^n-1까지의 모든 부분 집합을 생성
    for i in range(2**n):
        # subset의 초기 값을 0으로 설정
        subset = [0]*n
        # i를 2진수로 변환하여 부분 집합을 구함
        for j in range(n):
            # i를 이진수로 변환했을 때, 각 자리의 수를 리스트에 역순으로 저장
            # 예를 들어, i=5일 때, 이진수로 변환하면 101이고, 이를 역순으로 저장하면 [1,0,1]이 됨
            subset[j] = i%2
            i = i//2
            print(subset)
        # 부분 집합의 무게와 가치를 계산
        total_weight = 0
        total_value = 0
        for d in range(n):
            if subset[d] == 1:
                total_weight += weights[d]
                total_value += values[d]
        # 무게가 제한을 넘지 않고, 가치가 최대인 경우 갱신
        if total_weight <= W:
            if total_value > best_value:
                best_value = total_value
    return best_value


# 테스트 데이터
weights = [10, 20, 30, 25, 35]  # 물건별 무게
values = [60, 100, 120, 70, 85]  # 물건별 가치
W = 80  # 배낭 용량

# 실행
result = knapsack01_bruteforce(weights, values, W)
print("최대 가치:", result)
```
