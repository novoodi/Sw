```python
# 코드 2.3: 링버퍼를 위한 원형큐 삽입 수정 및 테스트 프로그램

class ArrayQueue :
    def __init__( self, capacity = 10 ) :
        self.capacity = capacity
        self.array = [None] * capacity
        self.front = 0
        self.rear = 0

    def isEmpty( self ) :
       return self.front == self.rear

    def isFull( self ) :
       return self.front == (self.rear+1) % self.capacity

    def enqueue( self, item ):
        if not self.isFull():
            self.rear = (self.rear + 1) % self.capacity
            self.array[self.rear] = item
        else : pass


    def dequeue(self):
          if not self.isEmpty():
              self.front = (self.front + 1) % self.capacity
              item = self.array[self.front]
              self.array[self.front] = None # 삭제된 자리 초기화
              return item
          else:
            print("Queue is empty. Cannot dequeue.")
            return None

    def peek( self ):
        if not self.isEmpty():
            return self.array[(self.front + 1) % self.capacity]
        else : pass

    def size( self ) :
       return (self.rear - self.front + self.capacity) % self.capacity

    def display(self, msg='Queue:' ):
        print(msg, end='= [')

        for i in range(self.front+1, self.front+1+self.size()):
            print(self.array[i % self.capacity], end=' ')
        print("]")


#########################추가 ######################
    # 코드 2.3a: 원형 큐: 링 버퍼를 위한 삽입 연산
    def enqueue2(self, item):
        self.rear = (self.rear + 1) % self.capacity
        self.array[self.rear] = item
        if self.isFull():
            self.front = (self.front + 1) % self.capacity # overwriting

#=========================================================
# 코드 2.3b: 링 버퍼의 테스트 프로그램

if __name__ == "__main__":
    q = ArrayQueue(8)
    q.display("초기 상태")

    for i in range(6) :
        q.enqueue2(i)
        print(f"queue:{q.array}")
        print(f"rear: {q.rear}")
        print(f"front: {q.front}")
        print()

    q.display("삽입 0-5")
    print()

    # enqueue2(): 6, 7
    q.enqueue2(6)
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()

    q.enqueue2(7)
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()

    q.display("삽입 6,7")
    print()

    # enqueue2(): 8, 9
    q.enqueue2(8)
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()
    q.enqueue2(9)
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()

    q.display("삽입 8,9")
    print()

    # dequeue() 2회
    q.dequeue()
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()
    q.dequeue()
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()

    q.display("삭제  x2")


```

# 1. 단순 연결 리스트 (singly linked list) 자료구조

```python
class Node:
    def __init__(self, elem, next = None):
        self.data = elem
        self.link = next

    def append(self, node):
        if node is not None:
            node.link = self.link # (1)
            self.link = node # (2)

    def popNext(self):
        next = self.link # 삭제 노드 정의
        if next is not None:
            self.link = next.link
        return next

class LinkedList:
    def __init__(self):
        self.head = None # 비어 있는 단순연결 리스트

    def isEmpty(self):
        return self.head == None

    def isFull(self):
        return False

    def getNode(self, pos):
    # 리스트에서 특정 위치에 있는 노드를 반환. 첫 번째 노드는 pos = 0
        if pos  < 0: return None

        ptr = self.head
        for i in range(pos):
            if ptr == None: # pos가 list의 크기보다 큰 경우
                return None
            ptr = ptr.link # ptr을 이동 진행

        return ptr # pos 위치에 있는 최종 노드을 반환

    def getEntry(self, pos):
        """
        리스트의 특정 위치에 있는 노드의 데이터 가져오기
        1. getNode() 함수를 이용하여 해당 위치의 노드을 가져온다.
        2. 그 노드의 data을 반환하고, 없으면 None을 반환
        """
        node = self.getNode(pos) # pos 위치에 있는 노드을 얻어오기
        if node is None:
            return None
        else:
            return node.data

    def replace(self,pos, elem):
        """
        리스트의 특정 위치에 있는 노드의 데이터을 변경
        1. getNode(pos)을 사용하여 해당 노드 얻어오기
        2. 그 노드의 data을 새로운 값으로 수정
        """
        node = self.getNode(pos)
        if node is None:
            return None
        else:
            node.data = elem

    def size(self):
        """
        연결리스트의 총 노드의 개수를 반환
        1. head부터 시작하여 모든 노드를 순차적으로 탐색하면서 카운트 한다.
        """
        ptr = self.head # 처음 노드부터 시작한다.
        count = 0

        while ptr is not None: #ptr이 None이 아닌 동안
            ptr = ptr.link # 링크를 따라 이동
            count+= 1 # 이동할 때마다 카운트 업데이트
        return count

    def display(self, msg = "LinkedList: "):
        """
        1. head부터 시작해서 각 노드의 data를 출력하며,
        2. 마지막 노드 이후에는 None을 출력
        """
        print(msg, end = '')
        ptr = self.head
        while ptr is not None:
            print(ptr.data, end = '->')
            ptr = ptr.link
        print("None")

    def insert(self, pos, elem):
        """
        특정한 위치에 새 노드를 삽입
        1. 저장할 새 노드를 생성한다.
        2. getNode(pos-1)을 사용하여 삽입할 위치의 이전 노드을 찾는다.
        2-1.  pos = 0이면 :  만약 위치가 리스트의 맨 앞일 경우, 새 노드를 head로 설정
        2-2  만약에 그렇지 않으면 그 앞에 새 노드를 삽입한다.
        """
        node = Node(elem, None)
        before = self.getNode(pos-1)
        if before == None : # pos = 0
            node.link = self.head
            self.head = node
        else:
            before.append(node)

    def delete(self, pos):
        """
        리스트에서 특정 위치의 노드를 삭제
        1. getNode()을 사용하여 삭제할 노드의 앞 노드를 찾는다.
        2. 만약 첫 번째 노드를 삭제하려면, self.head를 다음 노드로 변경.
        3. 그렇지 않는 경우, popNext()을 이용하여 해당 노드을 리스트에서 제외시킨다.
        """
        before = self.getNode(pos-1)
        if before == None:
            if self.head is not None: # 리스트가 공백이 아니라면
                self.head = self.head.link # 머리 노드을 갱신
        else: before.popNext()
```
