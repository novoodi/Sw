```python
# 코드 2.3: 링버퍼를 위한 원형큐 삽입 수정 및 테스트 프로그램

class ArrayQueue :
    def __init__( self, capacity = 10 ) :
        self.capacity = capacity
        self.array = [None] * capacity
        self.front = 0
        self.rear = 0

    def isEmpty( self ) :
       return self.front == self.rear

    def isFull( self ) :
       return self.front == (self.rear+1) % self.capacity

    def enqueue( self, item ):
        if not self.isFull():
            self.rear = (self.rear + 1) % self.capacity
            self.array[self.rear] = item
        else : pass


    def dequeue(self):
          if not self.isEmpty():
              self.front = (self.front + 1) % self.capacity
              item = self.array[self.front]
              self.array[self.front] = None # 삭제된 자리 초기화
              return item
          else:
            print("Queue is empty. Cannot dequeue.")
            return None

    def peek( self ):
        if not self.isEmpty():
            return self.array[(self.front + 1) % self.capacity]
        else : pass

    def size( self ) :
       return (self.rear - self.front + self.capacity) % self.capacity

    def display(self, msg='Queue:' ):
        print(msg, end='= [')

        for i in range(self.front+1, self.front+1+self.size()):
            print(self.array[i % self.capacity], end=' ')
        print("]")


#########################추가 ######################
    # 코드 2.3a: 원형 큐: 링 버퍼를 위한 삽입 연산
    def enqueue2(self, item):
        self.rear = (self.rear + 1) % self.capacity
        self.array[self.rear] = item
        if self.isFull():
            self.front = (self.front + 1) % self.capacity # overwriting

#=========================================================
# 코드 2.3b: 링 버퍼의 테스트 프로그램

if __name__ == "__main__":
    q = ArrayQueue(8)
    q.display("초기 상태")

    for i in range(6) :
        q.enqueue2(i)
        print(f"queue:{q.array}")
        print(f"rear: {q.rear}")
        print(f"front: {q.front}")
        print()

    q.display("삽입 0-5")
    print()

    # enqueue2(): 6, 7
    q.enqueue2(6)
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()

    q.enqueue2(7)
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()

    q.display("삽입 6,7")
    print()

    # enqueue2(): 8, 9
    q.enqueue2(8)
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()
    q.enqueue2(9)
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()

    q.display("삽입 8,9")
    print()

    # dequeue() 2회
    q.dequeue()
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    q.display("삭제  x2")
    print()
    q.dequeue()
    print(f"queue:{q.array}")
    print(f"rear: {q.rear}")
    print(f"front: {q.front}")
    print()
    q.display("삭제  x2")


```

# 1. 단순 연결 리스트 (singly linked list) 자료구조

```python
class Node:
    def __init__(self, elem, next=None):
        self.data = elem # elem에는 노드에 저장할 데이터가 들어감 
        self.link = next # 링크에는 다음 노드를 가리키는 링크가 저장 

    def append(self, node):
        if node is not None:
            # 새로 들어온 노드는 이전의 self의 링크가 가리키던 값을 가리김
            node.link = self.link
            # 이후 self의 링크는 새로 들어온 노드를 가리킴 
            self.link = node

    def popNext(self):
        # next는 self의 링크가 가리키는 값을 의미한다. 
        # 빈 값인지 확인하는 것 
        next = self.link
        if next is not None:
            # 빈 값이 아니라면 self의 링크가 가리키는 곳을 next의 링크
            # self의 링크의 링크를 가리킨다. 다다음 값 
            self.link = next.link
        # next를 반환 해줌으로써 어떤 값을 뺏는지 알려줌 
        return next


class LinkedList:
    def __init__(self):
        self.head = None  # 비어있는 단순 연결 리스트 아무것도 없는 상태
    def isEmpty(self):
        return self.head == None  # '=='로 비교 연산

    def isFull(self):
        return False

    def getNode(self, pos):
        # 리스트에서 특정 위치에 있는 노드를 반환, 첫번째 노드는 pos = 0
        # 0보다 작은 경우에는 오류이다.
        if pos < 0:
            return None
        ptr = self.head
        for i in range(pos):
            # 원하는 위치의 크기만큰 돌린다
            # 만약 3번위치를 원하면 ptr.link.link가 된다.
            if ptr is None:  # pos가 list의 크기보다 큰 경우 -> 맨 마지막 값은 None을 가리키기 때문에
                return None  # 'return None'으로 수정
            ptr = ptr.link  # ptr을 이동진행

        return ptr  # pos의 위치에 있는 최종 노드를 반환

    def getEntry(self, pos):
        """
        리스트의 특정 위치에 있는 노드의 데이터를 가져오기
        1. getNode()함수를 이용하여 해당 위치의 노드를 가져온다
        2. 그 노드의 data를 반환하고, 없으면 None으로 반환
        """
        node = self.getNode(pos)
        if node is None:
            return None
        else:
            # 해당 위치에 있는 노드의 데이터 값을 반환
            return node.data

    def replace(self, pos, elem):
        """
        리스트의 특정 위치에 있는 노드의 데이터를 변경
        1. getNode(pos)를 사용하여 해당 노드 얻어 오기
        2. 그 노드의 data를 새로운 값으로 수정
        """
        node = self.getNode(pos)
        if node is None:
            return None
        else:
            # 해당 위치에 있는 값을 원하는 값으로 변경
            node.data = elem

    def size(self):
        """
        연결리스트의 총 노드의 개수를 반환
        1. head부터 시작하여 모든 노드를 순차적으로 탐색하여 카운트한다.
        """
        ptr = self.head  # 처음 노드부터 시작한다.
        count = 0
        while ptr is not None:  # ptr이 None이 아닌 동안
            ptr = ptr.link  # 링크를 따라 이동
            count += 1  # 이동할 때마다 카운트를 업데이트
        return count

    def display(self, msg="Linked List: "):  # 'sef'를 'self'로 수정
        """
        1. head부터 시작해서 각 노드의 data를 출력
        2. 마지막 노드 이후에는 None을 출력
        """
        print(msg, end=" ")
        ptr = self.head
        while ptr is not None:
            print(ptr.data, end="->")
            ptr = ptr.link
        print("None")

    def insert(self, pos, elem):
        """
        특정한 위치에 새로운 노드를 삽입
        1. 저장할 새로운 노드를 생성한다.
        2. getNode(pos-1)을 사용해서 삽입할 위치의 이전 노드를 찾는다.
        2-1. pos=0이면 -> 위치가 맨 앞일 경우, 새로운 노드를 head로 설정하는 것
        2-2. 만약 그렇지 않다면 그 앞에 새로운 노드를 삽입한다.
        """
        node = Node(elem, None)
        before = self.getNode(pos - 1)
        if before is None:
            node.link = self.head
            self.head = node
        else:
            before.append(node)  # 'noe'를 'node'로 수정

    def delete(self, pos):
        """
        리스트에서 특정 위치의 노드를 삭제
        1. getNode()를 사용해서 삭제할 노드의 앞 노드를 찾는다.
        2. 만약 첫번째 노드를 삭제하려면, self.head를 다음 노드로 변경.
        3. 만약 그렇지 않다면 popNext()를 이용하여 해당 노드를 리스트에서 제외시킨다.
        """
        before = self.getNode(pos - 1)
        if before is None:
            before = self.head
            if self.head is not None:  # 리스트가 공백이 아니라면
                self.head = self.head.link  # 머리 노드를 갱신
            return before
        else:
            return before.popNext()
```

# 2. 이중 연결 리스트

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def isEmpty(self):
        return self.head is None

    def insert(self, pos, elem):
        new_node = Node(elem)
        if pos == 0:
            # Insert at the beginning
            if self.isEmpty():
                self.head = new_node
                self.tail = new_node
            else:
                new_node.next = self.head
                self.head.prev = new_node
                self.head = new_node
        else:
            # Insert at a given position
            prev_node = self.getNode(pos - 1)
            if prev_node is None:
                print(f"Position {pos} is out of bounds.")
                return
            next_node = prev_node.next
            new_node.prev = prev_node
            new_node.next = next_node
            prev_node.next = new_node
            if next_node:
                next_node.prev = new_node
            else:
                # If next_node is None, new_node is the new tail
                self.tail = new_node

    def delete(self, pos):
        if self.isEmpty():
            print("List is empty. Cannot delete.")
            return None
        if pos == 0:
            # Delete the head
            removed = self.head
            if self.head == self.tail:
                # Only one element in the list
                self.head = None
                self.tail = None
            else:
                self.head = self.head.next
                self.head.prev = None
            return removed.data
        else:
            # Delete at a given position
            prev_node = self.getNode(pos - 1)
            if prev_node is None or prev_node.next is None:
                print(f"Position {pos} is out of bounds.")
                return None
            removed = prev_node.next
            next_node = removed.next
            prev_node.next = next_node
            if next_node:
                next_node.prev = prev_node
            else:
                # If next_node is None, we're deleting the tail
                self.tail = prev_node
            return removed.data

    def getNode(self, pos):
        # Helper function to get the node at a specific position
        if pos < 0:
            return None
        ptr = self.head
        for _ in range(pos):
            if ptr is None:
                return None
            ptr = ptr.next
        return ptr

    def display(self, msg="DoublyLinkedList:"):
        print(msg, end=" ")
        ptr = self.head
        while ptr is not None:
            print(ptr.data, end=" <-> ")
            ptr = ptr.next
        print("None")

    def reverseDisplay(self, msg="DoublyLinkedList (Reverse):"):
        print(msg, end=" ")
        ptr = self.tail
        while ptr is not None:
            print(ptr.data, end=" <-> ")
            ptr = ptr.prev
        print("None")

# 사용 예시
dll = DoublyLinkedList()

data = dll.delete(0)
print(data)  # 리스트가 비어있으므로 None 출력
dll.display()

dll.insert(0, 10)  # 첫 번째 위치에 10 삽입
print(dll.head.data)  # 출력: 10
print(dll.tail.data)  # 출력: 10
dll.insert(1, 20)  # 두 번째 위치에 20 삽입
dll.insert(2, 30)  # 세 번째 위치에 30 삽입
dll.insert(1, 15)  # 두 번째 위치에 15 삽입

dll.display()  # 출력: DoublyLinkedList: 10 <-> 15 <-> 20 <-> 30 <-> None

dll.delete(2)  # 세 번째 위치의 노드 삭제 (20 삭제)
dll.display()  # 출력: DoublyLinkedList: 10 <-> 15 <-> 30 <-> None

dll.reverseDisplay()  # 출력: DoublyLinkedList (Reverse): 30 <-> 15 <-> 10 <-> None
```
