```python
# 코드 1.7: 반복 구조의 팩토리얼 함수
# 누적의 곱
def fact_iter(n) :
    # 지역변수 result
    result = 1
    for k in range(2, n+1) :	# k: 2, ..., n
        result = result * k	# result에 반복적으로 곱함
    return result


# 코드 1.8: 순환 구조의 팩토리얼 함수
# 재귀함수 
def fact_recur(n) :		# 순환적으로 구현한 factorial 함수
    if n == 1 : return 1		# 종료 조건: 순환을 멈추는 부분
    else :
        return n * fact_recur(n - 1)	# 자신을 순환적으로 호출
def main():
    number = int(input("양수를 입력해주세여: "))
    print(f"{number}!={fact_iter(number)}")
    print(f"{number}!={fact_recur(number)}")
if  __name__=="__main__":
    main()
```

# Fibonacci sequence 에서 n번째 항의 값 계산하기

```python
# 반복
def fibonacci_interative(n):
    if n <+0:
        return 0
    elif n ==1:
        return 1
    prev, current = 0,1
    for _ in range(2, n + 1):
        prev, current = current, prev + current
    return current
# 순환 
def fibonacci_recursive(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_recursive(n -1) + fibonacci_recursive(n-2)
def main():
    index = int(input("피보나치 수열에서 구할 항을 입력하세요 : "))
    print(f"{index}={fibonacci_interative(index)}")
    print(f"{index}={fibonacci_recursive(index)}")
if __name__ == "__main__":
    main()
```

```python
# 원형 큐 구현: 배열 방식 (전역 변수 사용)

# 전역 변수 선언
array = []
front = 0
rear = 0
capacity = 0

def create_queue(c=10): # 큐 생성의 초기화 
    global array, front, rear, capacity
    capacity = c
    array = [None] * capacity  # 순차 구조: 용량에 맞는 배열 생성
    front = 0  # 전단 인덱스 초기화
    rear = 0   # 후단 인덱스 초기화

def is_empty(): # 공백 상태
    global front, rear # 전역 변수 
    return front == rear# 0 or 1일 때 리턴 

def is_full(): # 포화 상태 
    global front, rear, capacity
    return front == (rear + 1) % capacity  # 원형 큐에서는 후단이 전단 바로 앞에 있으면 포화 상태

def enqueue(item):
    global array, rear, front, capacity
    if not is_full():  # 포화 상태가 아닌 경우
        array[rear] = item
        rear = (rear + 1) % capacity  # 후단 인덱스를 순환하도록 처리
    else:
        print("Queue is full. Cannot enqueue.")

def dequeue():
    global array, front, rear, capacity
    if not is_empty():  # 공백 상태가 아닌 경우
        item = array[front]
        front = (front + 1) % capacity  # 전단 인덱스를 순환하도록 처리
        array[front] = None # 삭제된 조기화 자라
        return item
    else:
        print("Queue is empty. Cannot dequeue.")
        return None

def peek():
    global array, front, capacity
    if not is_empty():  # 공백 상태가 아닌 경우
        return array[front]
    else:
        print("Queue is empty. Nothing to peek.")
        return None

def size(): #전체 요소의 수를 확인하는 것
    global front, rear, capacity
    return (rear - front + capacity) % capacity  # 원형 큐에서 요소의 개수를 계산

def display(msg='Queue:'):
    global array, front, rear, capacity
    print(msg, end=' = [')
    i = front
    while i != rear:
        print(array[i], end=' ')
        i = (i + 1) % capacity  # 인덱스를 순환하도록 처리
    print(']')

# main 함수를 통해 동작 테스트
def main():
    # 큐 생성
    create_queue(5)  # 용량 5인 원형 큐 생성
    print("큐가 생성되었습니다.")
    display()

    # 요소 삽입 (enqueue)
    print("\n요소 삽입 (enqueue) 작업을 수행합니다.")
    enqueue(10)
    enqueue(20)
    enqueue(30)
    display()

    # 큐의 크기 확인
    print("\n큐의 현재 크기:", size()) #aueue의 크가가 3으로 나와야함 

    # 요소 삭제 (dequeue)
    print("\n요소 삭제 (dequeue) 작업을 수행합니다.")
    dequeue()
    display()

    # 큐의 상단 요소 확인 (peek)
    print("\n상단 요소 확인 (peek):", peek())

    # 요소 추가 삽입 (enqueue)
    print("\n다시 요소 삽입 (enqueue) 작업을 수행합니다.")
    enqueue(40)
    enqueue(50)
    enqueue(60)  # 큐가 포화 상태일 경우 메시지 출력
    display()

    # 큐가 포화 상태인지 확인
    print("\n큐가 포화 상태인가?:", is_full())

    # 큐의 크기 확인
    print("\n큐의 현재 크기:", size())

    # 모든 요소 삭제
    print("\n모든 요소를 삭제합니다.")
    dequeue()
    dequeue()
    dequeue()
    dequeue()  # 큐가 공백 상태일 경우 메시지 출력
    display()

    # 큐가 공백 상태인지 확인
    print("\n큐가 공백 상태인가?:", is_empty())

# main 함수 호출
if __name__ == "__main__":
    main()
```
